```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for population growth
r = 0.15       # intrinsic growth rate
P0 = 5         # initial population

# Parameters for the dynamic carrying capacity
M1 = 2000        # initial carrying capacity
M2 = 1000        # additional increase, so final capacity = M1 + M2 = 70
a = 0.1        # steepness of increase in carrying capacity
t0 = 100       # inflection point (time at which increase is fastest)

# Time settings for simulation
t_start = 0
t_end = 200
dt = 0.1
time = np.arange(t_start, t_end + dt, dt)

# Arrays to store population and carrying capacity over time
P = np.zeros_like(time)
M = np.zeros_like(time)

# Initial condition
P[0] = P0

# Define the dynamic carrying capacity function M(t)
def carrying_capacity(t):
    return M1 + M2 / (1 + np.exp(-a * (t - t0)))

# Numerical integration using Euler's method
for i in range(1, len(time)):
    t_current = time[i-1]
    # Evaluate the carrying capacity at the current time
    M_current = carrying_capacity(t_current)
    M[i-1] = M_current  # store value for plotting
    # Compute the derivative dP/dt based on logistic growth with dynamic M(t)
    dPdt = r * P[i-1] * (1 - P[i-1] / M_current)
    # Euler update for population
    P[i] = P[i-1] + dt * dPdt

# Compute carrying capacity at the final time point
M[-1] = carrying_capacity(time[-1])

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(time, P, label="Population P(t)")
plt.plot(time, M, label="Carrying Capacity M(t)", linestyle="--")
plt.xlabel("Time")
plt.ylabel("Population")
plt.title("Population Growth with Dynamic Carrying Capacity")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


pop_bts = 20
pop_wts = 20
birthrate_bts = 1
birthrate_wts = 1
deathrate_bts = 0.0020 
deathrate_wts = 0.0027 
dt = 0.1 

y_pop_bts = []
y_pop_wts = []
x_time = []

for t in range(50):
    birth_bts = birthrate_bts * pop_bts * dt
    birth_wts = birthrate_wts * pop_wts * dt
    death_bts = deathrate_bts * pop_bts * pop_wts * dt 
    death_wts = deathrate_wts * pop_wts * pop_bts * dt 

    pop_bts = max(pop_bts + (birth_bts - death_bts), 0)  # Ensure no negative population
    pop_wts = max(pop_wts + (birth_wts - death_wts), 0)

    y_pop_bts.append(pop_bts)
    y_pop_wts.append(pop_wts)
    x_time.append(t)

    print(f"At time {t}, Population BTS: {pop_bts:.2f}")
    print(f"At time {t}, Population WTS: {pop_wts:.2f}")

# Plot results
plt.figure(figsize=(8, 5))
plt.plot(x_time, y_pop_bts, label="Population BTS", marker='o', color='red')
plt.plot(x_time, y_pop_wts, label="Population WTS", marker='s', linestyle="dashed", color='blue')
plt.xlabel("Time")
plt.ylabel("Population")
plt.title("Population Growth Over Time with Competition")
plt.legend()
plt.grid()
plt.show()
```
