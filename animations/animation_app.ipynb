```python
%%manim -qm -v Warning TwoTransforms

class TwoTransforms(Scene):
    def transform(self):
        a = Circle()
        b = Square()
        c = Triangle()
        self.play(Transform(a, b))
        self.play(Transform(a, c))
        self.play(FadeOut(a))

    def replacement_transform(self):
        a = Circle()
        b = Square()
        c = Triangle()
        self.play(ReplacementTransform(a, b))
        self.play(ReplacementTransform(b, c))
        self.play(FadeOut(c))

    def construct(self):
        self.transform()
        self.wait(0.5)  # wait for 0.5 seconds
        self.replacement_transform()
```

```python
%%manim -qm -v Warning MobjectExample

from manim import *

class MobjectExample(Scene):
    def construct(self):
        p1 = [-1,-1, 0]
        p2 = [ 1,-1, 0]
        p3 = [ 1, 1, 0]
        p4 = [-1, 1, 0]
        a  = Line(p1,p2).append_points(Line(p2,p3).points).append_points(Line(p3,p4).points)
        point_start  = a.get_start()
        point_end    = a.get_end()
        point_center = a.get_center()
        self.add(Text(f"a.get_start() = {np.round(point_start,2).tolist()}", font_size=24).to_edge(UR).set_color(YELLOW))
        self.add(Text(f"a.get_end() = {np.round(point_end,2).tolist()}", font_size=24).next_to(self.mobjects[-1],DOWN).set_color(RED))
        self.add(Text(f"a.get_center() = {np.round(point_center,2).tolist()}", font_size=24).next_to(self.mobjects[-1],DOWN).set_color(BLUE))

        self.add(Dot(a.get_start()).set_color(YELLOW).scale(2))
        self.add(Dot(a.get_end()).set_color(RED).scale(2))
        self.add(Dot(a.get_top()).set_color(GREEN_A).scale(2))
        self.add(Dot(a.get_bottom()).set_color(GREEN_D).scale(2))
        self.add(Dot(a.get_center()).set_color(BLUE).scale(2))
        self.add(Dot(a.point_from_proportion(0.5)).set_color(ORANGE).scale(2))
        self.add(*[Dot(x) for x in a.points])
        self.add(a)
```

```python
%%manim -qm -v Warning SinWaveFastSimulation

# from manim import *

class SinWaveFastSimulation(Scene):
    def construct(self):
        # Create axes
        axes = Axes(
            x_range=[0, 4 * PI, PI],
            y_range=[-1.5, 1.5, 0.5],
            axis_config={"color": WHITE},
        )

        # Add labels to axes
        # axes_labels = axes.get_axis_labels(x_label="x", y_label="y")
        # axes_labels = axes.get_axis_labels(x_label=MathTex("x").set_color(WHITE), y_label=MathTex("y").set_color(WHITE))
        axes_labels = VGroup()

        # Create sine wave
        sin_wave = axes.plot(
            lambda x: np.sin(x),
            x_range=[0, 4 * PI],
            color=BLUE,
        )

        # Create a dot to move along the wave
        moving_dot = Dot(color=RED).move_to(axes.c2p(0, 0))

        # Add vertical line from the dot to the x-axis
        vertical_line = always_redraw(
            lambda: axes.get_vertical_line(
                axes.c2p(moving_dot.get_x(), 0),
                color=YELLOW,
            )
        )

        # Animate the wave and the moving dot
        self.play(Create(axes), Write(axes_labels))
        self.play(Create(sin_wave))
        self.play(Create(moving_dot), Create(vertical_line))

        # Animation for moving the dot along the wave
        self.play(
            UpdateFromFunc(
                moving_dot,
                lambda m: m.move_to(
                    axes.c2p(moving_dot.get_x() + 0.1, np.sin(moving_dot.get_x() + 0.1))
                ),
            ),
            run_time=8,
            rate_func=linear,
        )

        self.wait()
```

```python
%%manim -ql -v Warning CountingScene



class Count(Animation):
    def __init__(self, number: DecimalNumber, start: float, end: float, **kwargs) -> None:
        # Pass number as the mobject of the animation
        super().__init__(number,  **kwargs)
        # Set start and end
        self.start = start
        self.end = end

    def interpolate_mobject(self, alpha: float) -> None:
        # Set value of DecimalNumber according to alpha
        value = self.start + (alpha * (self.end - self.start))
        self.mobject.set_value(value)


class CountingScene(Scene):
    def construct(self):
        # Create Decimal Number and add it to scene
        number = DecimalNumber().set_color(WHITE).scale(10)
        # Add an updater to keep the DecimalNumber centered as its value changes
        number.add_updater(lambda number: number.move_to(ORIGIN))

        self.add(number)

        self.wait()

        # Play the Count Animation to count from 0 to 100 in 4 seconds
        self.play(Count(number, 0, 2), run_time=50, rate_func=linear)

        self.wait()
```

```python
%%manim -qm -v Warning SinWaveSimulationCustom

from manim import *

class SinWaveSimulationCustom(Scene):
    def construct(self):
        # Create axes
        axes = Axes(
            x_range=[0, 4 * PI, PI],
            y_range=[-1.5, 1.5, 0.5],
            axis_config={"color": WHITE},
        )
        # Add labels to axes
        axes_labels = axes.get_axis_labels(x_label="x", y_label="y")

        # Create sine wave
        sin_wave = axes.plot(
            lambda x: np.sin(x),
            x_range=[0, 4 * PI],
            color=BLUE,
        )

        # Create a dot to move along the wave
        moving_dot = Dot(color=RED).move_to(axes.c2p(0, 0))

        # Add vertical line from the dot to the x-axis
        vertical_line = always_redraw(
            lambda: axes.get_vertical_line(
                axes.c2p(moving_dot.get_x(), 0),
                color=YELLOW,
            )
        )

        # Function to update the dot position dynamically
        def update_dot(mob, dt):
            current_x = mob.get_x() + dt  # Increment x-coordinate over time
            current_y = np.sin(current_x)  # Get the y-coordinate from the sine function
            mob.move_to(axes.c2p(current_x, current_y))

        # Add updater to the dot
        moving_dot.add_updater(update_dot)

        # Animate the wave and the moving dot
        self.play(Create(axes), Write(axes_labels))
        self.play(Create(sin_wave))
        self.add(moving_dot, vertical_line)

        # Let the dot move along the wave for 15 seconds
        self.wait(15)

        # Remove the updater after animation
        moving_dot.remove_updater(update_dot)
        self.wait()
```

```python
%%manim -qm -v Warning FourierTransformAnimation

from manim import *

class FourierTransformAnimation(Scene):
    def construct(self):
        # Initialize the number of terms
        self.num_terms = 1  # Start with 1 term

        # Create axes for original square wave
        axes_square = Axes(
            x_range=[0, 2 * PI, PI / 2],
            y_range=[-1.5, 1.5, 0.5],
            axis_config={"color": WHITE},
            tips=False,
        ).to_edge(UP)

        # Create axes for Fourier reconstruction
        axes_fourier = Axes(
            x_range=[0, 2 * PI, PI / 2],
            y_range=[-1.5, 1.5, 0.5],
            axis_config={"color": WHITE},
            tips=False,
        ).to_edge(DOWN)

        # Labels for axes
        axes_square_label = axes_square.get_axis_labels(x_label="x", y_label="f(x)")
        axes_fourier_label = axes_fourier.get_axis_labels(x_label="x", y_label="f_approx(x)")

        # Create square wave
        square_wave = axes_square.plot(
            lambda x: 1 if (x % (2 * PI)) < PI else -1,
            color=YELLOW,
        )

        # Fourier reconstruction wave (start with 1 term)
        reconstructed_wave = always_redraw(
            lambda: axes_fourier.plot(
                lambda x: sum(
                    (4 / (PI * n)) * np.sin(n * x)
                    for n in range(1, self.num_terms * 2, 2)
                ),
                color=BLUE,
            )
        )

        # Display original square wave
        self.play(Create(axes_square), Write(axes_square_label))
        self.play(Create(square_wave))

        # Display Fourier reconstruction
        self.play(Create(axes_fourier), Write(axes_fourier_label))
        self.add(reconstructed_wave)

        # Animate Fourier series terms
        for i in range(2, 20):  # Add up to 19 terms
            self.num_terms = i
            self.play(
                Transform(
                    Text(f"Terms: {self.num_terms}", font_size=24).to_corner(UL).set_color(RED),
                    Text(f"Terms: {i}", font_size=24).to_corner(UL).set_color(RED),
                ),
                run_time=0.5,
            )
            self.wait(0.2)

        # Pause at the final reconstruction
        self.wait(2)
```

```python
%%manim -ql -v Warning MatrixTransformation

from manim import *

class MatrixTransformation(Scene):
    def construct(self):
        # Define the axes
        axes = NumberPlane(
            x_range=[-5, 5, 1],
            y_range=[-5, 5, 1],
            background_line_style={"stroke_color": BLUE_E, "stroke_width": 1},
        )
        self.play(Create(axes))

        # Define a vector
        original_vector = Arrow(start=ORIGIN, end=[2, 1, 0], color=YELLOW, buff=0)
        original_label = MathTex("\\vec{v}", font_size=24).next_to(original_vector, UP)

        # Display the vector
        self.play(Create(original_vector), Write(original_label))

        # Define the transformation matrix
        matrix = [[1, 2], [0, 1]]
        matrix_label = Matrix(matrix, left_bracket="(", right_bracket=")")
        matrix_label.to_edge(UP)

        # Show the matrix on the screen
        self.play(Write(matrix_label))

        # Calculate the transformed vector
        transformed_vector_coords = [
            matrix[0][0] * 2 + matrix[0][1] * 1,  # x' = 1*2 + 2*1
            matrix[1][0] * 2 + matrix[1][1] * 1,  # y' = 0*2 + 1*1
        ]
        transformed_vector = Arrow(
            start=ORIGIN,
            end=[transformed_vector_coords[0], transformed_vector_coords[1], 0],
            color=RED,
            buff=0,
        )
        transformed_label = MathTex(
            "\\vec{v}'", font_size=24
        ).next_to(transformed_vector, UP)

        # Animate the transformation
        self.play(Transform(original_vector, transformed_vector))
        self.play(Transform(original_label, transformed_label))

        # Add explanation text
        explanation = Text(
            "Matrix Transformation: \\vec{v} \\to \\vec{v}'", font_size=24
        )
        explanation.to_edge(DOWN)
        self.play(Write(explanation))

        self.wait(2)
```

```python
%%manim -ql -v Warning ColumnToRowTransformation

from manim import *

class ColumnToRowTransformation(Scene):
    def construct(self):
        # Create numbers (1 to 4) inside individual boxes
        column = VGroup(
            *[
                VGroup(Square(side_length=1, color=WHITE), Text(str(i)))
                for i in range(1, 6)
            ]
        )

        # Position the column vertically
        for idx, box in enumerate(column):
            box[1].move_to(box[0].get_center())  # Center the number inside the box
            box.shift(DOWN * idx)

        # Display the column
        self.play(FadeIn(column))

        # Transform the column into a row
        row = column.copy()
        row.arrange(RIGHT, buff=0.5)  # Arrange boxes horizontally

        # Animate the transformation
        self.play(Transform(column, row))

        self.wait()
```

```python
%%manim -ql -v Warning ThreeDAnimation

# from manim import *

class ThreeDAnimation(ThreeDScene):
    def construct(self):
        # Set up the 3D camera
        self.set_camera_orientation(phi=75 * DEGREES, theta=30 * DEGREES)

        # Create a cube
        cube = Cube(side_length=2, color=BLUE).shift(LEFT * 3)
        cube_text = Text("Cube").next_to(cube, DOWN)

        # Create a sphere
        sphere = Sphere(radius=1.5, color=GREEN).shift(RIGHT * 3)
        sphere_text = Text("Sphere").next_to(sphere, DOWN)

        # Add the cube and sphere to the scene
        self.play(Create(cube), Write(cube_text))
        self.play(Create(sphere), Write(sphere_text))

        # Rotate the cube
        self.play(Rotate(cube, angle=PI / 2, axis=RIGHT), run_time=2)
        self.play(Rotate(cube, angle=PI / 2, axis=UP), run_time=2)

        # Rotate the sphere
        self.play(Rotate(sphere, angle=PI, axis=RIGHT), run_time=2)
        self.play(Rotate(sphere, angle=PI, axis=UP), run_time=2)

        # Change camera orientation dynamically
        self.move_camera(phi=60 * DEGREES, theta=45 * DEGREES, run_time=3)
        self.wait(1)

        # Spin both objects simultaneously
        self.play(
            Rotate(cube, angle=2 * PI, axis=UP),
            Rotate(sphere, angle=2 * PI, axis=RIGHT),
            run_time=4,
        )
        self.wait()
```

```python
%%manim -ql -v Warning BoxArrowAnimation

from manim import *

class BoxArrowAnimation(Scene):
    def construct(self):
        # Step 1: Create the first box in the center and add text
        box1 = Square(side_length=2, color=BLUE)
        text1 = Text("Start").move_to(box1.get_center())
        group1 = VGroup(box1, text1)  # Group the box and text together

        # Animate the appearance of the first box and text
        self.play(Create(box1), Write(text1))
        self.wait(1)

        # Step 2: Move the first box to the top and shrink its size
        self.play(group1.animate.scale(0.5).to_edge(UP), run_time=2)
        self.wait(1)

        # Step 3: Draw an arrow pointing down
        arrow = Arrow(start=group1.get_bottom(), end=ORIGIN, color=YELLOW)

        # Step 4: Create the second box at the center with text after the arrow reaches
        box2 = Square(side_length=2, color=GREEN)
        text2 = Text("Next Step").move_to(box2.get_center())
        group2 = VGroup(box2, text2)

        # Animate the arrow and then the second box
        self.play(Create(arrow), run_time=1.5)
        self.play(FadeIn(group2), run_time=1.5)

        # Wait for a moment before ending
        self.wait(2)
```

```python
class TangentAnimation(Scene):
    def construct(self):
        ax = Axes()
        sine = ax.plot(np.sin, color=RED)
        alpha = ValueTracker(0)
        point = always_redraw(
            lambda: Dot(
                sine.point_from_proportion(alpha.get_value()),
                color=BLUE
            )
        )
        tangent = always_redraw(
            lambda: TangentLine(
                sine,
                alpha=alpha.get_value(),
                color=YELLOW,
                length=4
            )
        )
        self.add(ax, sine, point, tangent)
        self.play(alpha.animate.set_value(1), rate_func=linear, run_time=1)



# don't remove below command for run button to work
%manim -ql -v WARNING TangentAnimation
```

```python
class SpinningTriangle(Scene):
    def construct(self):
        tri = Triangle().set_fill(opacity=1).set_z_index(2)
        sq = Square().to_edge(LEFT)

        # will keep spinning while there is an animation going on
        always_rotate(tri, rate=0.1*PI, about_point=ORIGIN)

        self.add(tri, sq)
        self.play(sq.animate.to_edge(RIGHT), rate_func=linear, run_time=10)



# don't remove below command for run button to work
%manim -qm -v WARNING SpinningTriangle
```

```python
class ShiftingSquare(Scene):
    def construct(self):
        sq = Square().set_fill(opacity=1)
        tri = Triangle()
        VGroup(sq, tri).arrange(LEFT)

        # construct a square which is continuously
        # shifted to the right
        always_shift(sq, RIGHT, rate=1)

        self.add(sq)
        self.play(tri.animate.set_fill(opacity=1))



# don't remove below command for run button to work
%manim -ql -v WARNING ShiftingSquare
```

```python
class FollowingGraphCamera(MovingCameraScene):
    def construct(self):
        self.camera.frame.save_state()

        # create the axes and the curve
        ax = Axes(x_range=[-1, 10], y_range=[-1, 10])
        graph = ax.plot(lambda x: np.sin(x), color=BLUE, x_range=[0, 3 * PI])

        # create dots based on the graph
        moving_dot = Dot(ax.i2gp(graph.t_min, graph), color=ORANGE, radius=0.08)
        dot_1 = Dot(ax.i2gp(graph.t_min, graph))
        dot_2 = Dot(ax.i2gp(graph.t_max, graph))

        self.add(ax, graph, dot_1, dot_2, moving_dot)
        self.play(self.camera.frame.animate.scale(0.5).move_to(moving_dot))

        def update_curve(mob):
            mob.move_to(moving_dot.get_center())

        self.camera.frame.add_updater(update_curve)
        self.play(MoveAlongPath(moving_dot, graph, rate_func=linear))
        self.camera.frame.remove_updater(update_curve)

        self.play(Restore(self.camera.frame))



# don't remove below command for run button to work
%manim -qm -v WARNING FollowingGraphCamera
```

```python
from manim import *

class MoveAlongPathExample(Scene):
    def construct(self):
        d1 = Dot().set_color(ORANGE)
        l1 = Line(LEFT, RIGHT)
        l2 = VMobject()
        self.add(d1, l1, l2)
        l2.add_updater(lambda x: x.become(Line(LEFT, d1.get_center()).set_color(ORANGE)))
        self.play(MoveAlongPath(d1, l1), rate_func=linear)
        
# don't remove below command for run button to work
%manim -qm -v WARNING MoveAlongPathExample        
```

```python
from manim import *

class LaggedStartExample(Scene):
    def construct(self):
        title = Text("lag_ratio = 0.25").to_edge(UP)

        dot1 = Dot(point=LEFT * 2 + UP, radius=0.16)
        dot2 = Dot(point=LEFT * 2, radius=0.16)
        dot3 = Dot(point=LEFT * 2 + DOWN, radius=0.16)
        line_25 = DashedLine(
            start=LEFT + UP * 2,
            end=LEFT + DOWN * 2,
            color=RED
        )
        label = Text("25%", font_size=24).next_to(line_25, UP)
        self.add(title, dot1, dot2, dot3, line_25, label)

        self.play(LaggedStart(
            dot1.animate.shift(RIGHT * 4),
            dot2.animate.shift(RIGHT * 4),
            dot3.animate.shift(RIGHT * 4),
            lag_ratio=0.05,
            run_time=2
        ))
        
        self.play(LaggedStart(
            dot1.animate.shift(LEFT * 4),
            dot2.animate.shift(LEFT * 4),
            dot3.animate.shift(LEFT * 4),
            lag_ratio=0.05,
            run_time=2
        ))
        
%manim -qm -v WARNING LaggedStartExample         
```

```python
from manim import *

class LaggedStartMapExample(Scene):
    def construct(self):
        title = Tex("LaggedStartMap").to_edge(UP, buff=LARGE_BUFF)
        dots = VGroup(
            *[Dot(radius=0.16) for _ in range(35)]
            ).arrange_in_grid(rows=5, cols=7, buff=MED_LARGE_BUFF)
        self.add(dots, title)

        # Animate yellow ripple effect
        for mob in dots, title:
            self.play(LaggedStartMap(
                ApplyMethod, mob,
                lambda m : (m.set_color, YELLOW),
                lag_ratio = 0.1,
                rate_func = there_and_back,
                run_time = 2
            ))

%manim -qm -v WARNING LaggedStartMapExample                     
```
