## K-Means++

sklearn.cluster.kmeans_plusplus

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import kmeans_plusplus
from sklearn.datasets import make_blobs

n_samples = 4000
n_components = 4

X, y_true = make_blobs(n_samples=n_samples, centers=n_components, cluster_std=0.60, random_state=0)

```

```python
print(X[1:5],'\n', y_true[:10])
```

```python
X = X[:, ::-1]
```

```python
X
```

```python
center_int, indices =  kmeans_plusplus(X, n_clusters=4, random_state=0)
```

```python
plt.figure(1)
colors = ["#4EACC5", "#FF9C34", "#4E9A06", "m"]

for k, col in enumerate(colors):
    cluster_data = y_true == k
    plt.scatter(X[cluster_data,0],X[cluster_data,1], c=col, marker=".", s=10)

plt.scatter(center_int[:,0], center_int[:,1], c="b", s=50)
plt.title("K-Means++ Initialization")
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
plt.figure(1)
colors = ["#4EACC5", "#FF9C34", "#4E9A06", "m"]

for k, col in enumerate(colors):
    cluster_data = y_true == k
    plt.scatter(X[cluster_data,0],X[cluster_data,1], c=col, marker="*", s=0.5)

plt.scatter(center_int[:,0], center_int[:,1], c="b", s=50)
plt.title("K-Means++ Initialization")
plt.xticks([])
plt.yticks([])
plt.show()
```
