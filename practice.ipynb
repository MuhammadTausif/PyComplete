```python
import curses

def maina(stdscr):
    # Clear screen
    stdscr.clear()

    # This raises an exception if there are not enough colors supported by the terminal.
    curses.start_color()
    curses.init_pair(1, curses.COLOR_RED, curses.COLOR_WHITE)

    # Print a colored string
    stdscr.addstr("Hello, World!", curses.color_pair(1))
    stdscr.refresh()

    # Wait for user input
    stdscr.getkey()

curses.wrapper(maina)
```
