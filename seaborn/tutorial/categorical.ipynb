# Categorical

```python
import seaborn as sns
tips = sns.load_dataset("tips")
```

```python
sns.catplot(data=tips, x="day", y="total_bill")
```

```python
sns.catplot(data=tips, x="day", y="total_bill", hue="smoker", kind="box")
```

```python
sns.catplot(data=tips, x="day", y="total_bill", kind="swarm")
```

### Kind swarm

```python
sns.catplot(data=tips, x="day", y="total_bill", kind="swarm")
```

### Kind=swarm and hue=sex

```python
sns.catplot(data=tips, x="day", y="total_bill", hue="sex", kind="swarm")
```

### Use query to filter size

```python
sns.catplot(data=tips.query("size != 3"), x="size", y="total_bill")
```

```python
sns.catplot(data=tips.query("size != 3"), x="size", y="total_bill", kind = "swarm")
```

```python
sns.catplot(data=tips.query("size != 3"), x="size", y="total_bill", hue="sex", kind="swarm")
```

### native_scale = True

```python
sns.catplot(data=tips.query("size != 3"), x="size", y="total_bill", native_scale=True)
```

```python
sns.catplot(data=tips.query("size != 3"), x="size", y="total_bill", native_scale=True, kind="swarm")
```

### Order

```python
sns.catplot(data=tips, x="smoker", y="tip", order=["No", "Yes"], kind="swarm")
```

```python
sns.catplot(data=tips, x="smoker", y="tip", order=["No", "Yes"], kind="bar")
```

```python
sns.catplot(data=tips, x="smoker", y="tip", order=["No", "Yes"], kind="box")
```

```python
sns.catplot(data=tips, x="smoker", y="tip", order=["No", "Yes"], kind="boxen")
```

```python
sns.catplot(data=tips, x="smoker", y="tip", order=["No", "Yes"], kind="point")
```

```python
sns.catplot(data=tips, x="smoker", y="tip", order=["No", "Yes"], kind="strip")
```

```python
sns.catplot(data=tips, x="smoker", y="tip", order=["No", "Yes"], kind="violin")
```

```python
sns.catplot(data=tips, x="smoker", y="tip", order=["No", "Yes"], kind="violin", hue="sex")
```

### Kind = swarm, hue = time

```python
sns.catplot(data=tips, x="total_bill", y="day", hue="time", kind="swarm")
```

```python
sns.catplot(data=tips, x="total_bill", y="day", hue="time", kind="violin")
```

## Comparing distributions

### Boxplots

```python
sns.catplot(data=tips, x="day", y="total_bill", kind="box")
```

#### Hue = smoker, kind=box

```python
sns.catplot(data=tips, x="day", y="total_bill", hue="smoker", kind="box")
```

#### hue = "weekend", tips = not weekend

```python
tips["weekend"] = tips["day"].isin(["Sat", "Sun"])
sns.catplot(data=tips, x="day", y="total_bill", hue="weekend", kind="box")
```

#### Dataset diomonds, kind=box, hue = weekend

```python
diamonds = sns.load_dataset("diamonds")
sns.catplot(
    data=diamonds.sort_values("color"),
    x="color", y="price", kind="boxen",
)
```
